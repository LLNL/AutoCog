```python
tree = wrap_frontend("""
prompt main {
    is {
        topic is text;
    }
    channel {
        to .choices call {
            extern another_cog;
            entry mmlu_annot;
            kwarg question from ?question;
            kwarg choice   map  ?choices;
            bind choice.test as value;
        }
    }
}
""")
dotstr = tree.toGraphViz()
# print(dotstr)
wrap_graphviz(dotstr)
```
